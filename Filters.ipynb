```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter

image = cv2.imread('/content/drive/My Drive/etik/snow2.jpg') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
figure_size = 9 
new_image = cv2.blur(image,(figure_size, figure_size))
plt.figure(figsize=(15,8))
plt.subplot(221), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Orjinal Resim')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB)),plt.title('Ortalama (Mean) Filtre')
plt.xticks([]), plt.yticks([])

image2 = cv2.cvtColor(image, cv2.COLOR_HSV2BGR)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
figure_size = 9
new_image2 = cv2.blur(image2,(figure_size, figure_size))
#plt.figure(figsize=(15,15))
plt.subplot(223), plt.imshow(image2, cmap='gray'),plt.title('Orjinal Resim')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(new_image2, cmap='gray'),plt.title('Ortalama (Mean) Filtre')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
new_image = cv2.medianBlur(image, figure_size)
plt.figure(figsize=(15,8))
plt.subplot(221), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Orjinal')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB)),plt.title('Medyan Filtre')
plt.xticks([]), plt.yticks([])

new_image = cv2.medianBlur(image2, figure_size)

plt.subplot(223), plt.imshow(image2, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(new_image, cmap='gray'),plt.title('Median Filtre')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
new_image = cv2.GaussianBlur(image, (figure_size, figure_size),0)
plt.figure(figsize=(15,8))
plt.subplot(221), plt.imshow(cv2.cvtColor(image, cv2.COLOR_HSV2RGB)),plt.title('Orjinal')
plt.xticks([]), plt.yticks([])
plt.subplot(222), plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_HSV2RGB)),plt.title('Gauss Filtresi')
plt.xticks([]), plt.yticks([])

new_image_gauss = cv2.GaussianBlur(image2, (figure_size, figure_size),0)

plt.subplot(223), plt.imshow(image2, cmap='gray'),plt.title('Orjinal')
plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.imshow(new_image_gauss, cmap='gray'),plt.title('Gauss Filtresi')
plt.xticks([]), plt.yticks([])
plt.show()
```
