```python
import random
from numpy.random import default_rng
import numpy as np

mux, sigmax = 0,1 # mean and standard deviation
muy, sigmay = 0,2
```

```python
import matplotlib.pyplot as plt
x = [] 
k = []

    
for i in range(150000): 
    temp = random.gauss(mux, sigmax)
    x.append(temp) 
stdx = np.std(x, ddof=0)
np.mean(x) 
np.mean(stdx)        
plt.plot(x)
plt.ylabel("Sigma Değer Aralığı")
plt.xlabel("Zaman")
plt.show()
print("standart sapma:" , stdx)
```

```python
for j in range(150000): 
    temp = random.gauss(muy, sigmay)
    k.append(temp) 
stdy = np.std(k, ddof=0)
np.mean(k)
np.mean(stdy)        
# plotting a graph

plt.plot(k, color='red')

plt.ylabel("Sigma Değer Aralığı")
plt.xlabel("Zaman")
plt.show()
print("standart sapma:" ,stdy)
```

```python
for j in range(150000): 
    temp = random.gauss(muy, sigmay)
    k.append(temp) 
for i in range(150000): 
    temp = random.gauss(mux, sigmax)
    x.append(temp)    
plt.figure(figsize=(10,6)) 
plt.plot(k, color='red')
plt.plot(x)
plt.ylabel("Sigma Değer Aralığı")
plt.xlabel("Zaman")
plt.show()
```

```python
num_binsx = 100


plt.figure(figsize=(12,6))
#fig, ax = plt.subplots()

n, bins, patches = plt.hist(x, num_binsx, density=1)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigmax)) *
     np.exp(-0.5 * (1 / sigmax * (bins - mux))**2))
plt.plot(bins, y, '--')

#ax.set_xlabel('Smarts')
#ax.set_ylabel('Probability density')
#plt.set_title(r'Histogram: standart sapma = 1, $\sigma=-3,+3$')
#fig.tight_layout()
plt.show()
```
